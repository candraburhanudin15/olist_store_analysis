##### Data cleaning orders_df

- mengkonversi tipe data

```python
datetimes_orders_df = ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date']
for column in datetimes_orders_df:
    orders_df[column]= pd.to_datetime(orders_df[column])
```

- drop data invalid values

karena terdapat kekeliruan data dimana tanggal order approved_at barang lebih lama daripada tanggal order_delivered_carrier_date lebih dahulu. data perlu dihapus

```python
data_shipped_abnormal = orders_df[(orders_df['order_status']== 'shipped') & \
                                  (orders_df['order_delivered_carrier_date'].notna()) & \
                                  (orders_df['order_approved_at'] > orders_df['order_delivered_carrier_date'])]
orders_df = orders_df.drop(data_shipped_abnormal.index)
```

```python
data_delivery_abnormal = orders_df[(orders_df['order_status']== 'delivered') & \
                                    (orders_df['order_delivered_carrier_date'].notna()) & \
                                    (orders_df['order_approved_at'] > orders_df['order_delivered_carrier_date'])]
orders_df = orders_df.drop(data_delivery_abnormal.index)
```

karena terdapat kekeliruan data dimana tanggal order_delivered_carrier_date barang lebih lama daripada tanggal order_delivered_customer_date lebih dahulu yang seharusnya kebalikannya. data perlu dihapus

```python
data_delivered_abnormal = orders_df[(orders_df['order_status']== 'delivered') & \
                                    (orders_df['order_delivered_customer_date'].notna()) & \
                                    (orders_df['order_delivered_carrier_date'] > orders_df['order_delivered_customer_date'])]
orders_df = orders_df.drop(data_delivered_abnormal.index)
```

- penanganan missing values pada order_status = unavailable dengan mengisi data tanggal terakhir (order_approved_at)

```python
orders_df[(orders_df['order_status'] == 'unavailable') & orders_df['order_approved_at'].notna()] 
```

```python
datetimes_orders_df_unavailable = ['order_delivered_carrier_date','order_delivered_customer_date']
for column in datetimes_orders_df_unavailable:
    mask = orders_df['order_status'] == 'unavailable'
    orders_df.loc[mask, column] = orders_df.loc[mask, column].fillna(orders_df.loc[mask, 'order_approved_at'])
```

- penanganan missing values pada order_status = processing dengan mengisi data tanggal terakhir (order_approved_at)

```python
orders_df[(orders_df['order_status'] == 'processing') & orders_df['order_approved_at'].notna()] 
```

```python
datetimes_orders_df_processing = ['order_delivered_carrier_date','order_delivered_customer_date']
for column in datetimes_orders_df_processing:
    mask = orders_df['order_status'] == 'processing'
    orders_df.loc[mask, column] = orders_df.loc[mask, column].fillna(orders_df.loc[mask, 'order_approved_at'])
```

- penanganan missing values pada order_status = invoiced dengan mengisi data tanggal terakhir (order_approved_at)

```python
orders_df[(orders_df['order_status'] == 'invoiced') & orders_df['order_approved_at'].notna()] 
```

```python
datetimes_orders_df_invoiced = ['order_delivered_carrier_date','order_delivered_customer_date']
for column in datetimes_orders_df_invoiced:
    mask = orders_df['order_status'] == 'invoiced'
    orders_df.loc[mask, column] = orders_df.loc[mask, column].fillna(orders_df.loc[mask, 'order_approved_at'])
```

- penanganan missing values pada order_status = created dengan mengisi data tanggal terakhir (order_purchase_timestamp)

```python
datetimes_orders_df_created = ['order_approved_at','order_delivered_carrier_date','order_delivered_customer_date']
for column in datetimes_orders_df_created:
    mask = orders_df['order_status'] == 'created'
    orders_df.loc[mask, column] = orders_df.loc[mask, column].fillna(orders_df.loc[mask, 'order_purchase_timestamp'])
```

- penanganan missing values pada order_status = approved dengan mengisi data tanggal terakhir (order_approved_at)

```python
datetimes_orders_df_approved= ['order_delivered_carrier_date','order_delivered_customer_date']
for column in datetimes_orders_df_approved:
    mask = orders_df['order_status'] == 'approved'
    orders_df.loc[mask, column] = orders_df.loc[mask, column].fillna(orders_df.loc[mask, 'order_approved_at'])
```

- penanganan missing values pada order_status = shipped dengan mengisi data tanggal terakhir (order_delivered_carrier_date)

```python
datetimes_orders_df_shipped= ['order_delivered_customer_date']
for column in datetimes_orders_df_shipped:
    mask = orders_df['order_status'] == 'shipped'
    orders_df.loc[mask, column] = orders_df.loc[mask, column].fillna(orders_df.loc[mask, 'order_delivered_carrier_date'])
```

- penanganan missing values pada order_status = canceled. dengan mengisi data tanggal terakhir (order_purchase_timestamp). kondisi ini mengelompokkan order dengan tingkat cancel paling awal (bisa terjadi karena cancel saat proses pembelian setelah pembayaran)

```python
datetimes_orders_df_early_cancellation = ['order_approved_at','order_delivered_carrier_date', 'order_delivered_customer_date']

for column in datetimes_orders_df_early_cancellation:
    mask = (orders_df['order_status'] == 'canceled') & \
           orders_df['order_approved_at'].isna() & \
           orders_df['order_delivered_carrier_date'].isna() & \
           orders_df['order_delivered_customer_date'].isna()

    orders_df.loc[mask, column] = orders_df.loc[mask, column] = orders_df.loc[mask, 'order_purchase_timestamp']
```

- penanganan missing values pada order_status = canceled. dengan mengisi data tanggal terakhir (order_approved_at). kondisi ini mengelompokkan order dengan tingkat cancel menengah (bisa terjadi proses order sudah di setujui oleh penjual namun pembeli melakukan cancel order)

```python
datetimes_orders_df_mid_cancellation = ['order_delivered_carrier_date', 'order_delivered_customer_date']
for column in datetimes_orders_df_mid_cancellation:
    mask = (orders_df['order_status'] == 'canceled') & \
            orders_df['order_delivered_carrier_date'].isna() & \
            orders_df['order_delivered_customer_date'].isna()
    
                        
    orders_df.loc[mask, column] = orders_df.loc[mask, column] = orders_df.loc[mask, 'order_approved_at']
```

- penanganan missing values pada order_status = canceled. dengan mengisi data tanggal terakhir (order_delivered_carrier_date). kondisi ini mengelompokkan order dengan tingkat cancel terakhir (bisa terjadi proses order barang sedang pengiriman namun pembeli melakukan proses cancel)

```python
datetimes_orders_df_late_cancellation = ['order_delivered_customer_date']
for column in datetimes_orders_df_mid_cancellation:
    mask = (orders_df['order_status'] == 'canceled') & \
            orders_df['order_delivered_customer_date'].isna()
    orders_df.loc[mask, column] = orders_df.loc[mask, column] = orders_df.loc[mask, 'order_delivered_carrier_date']
```

- mengisi beberapa kolom missing values (order_approved_at) dengan metode Imputation pada order yang berhasil (order_status = delivered) dengan menggunakan nilai rata rata pada setiap kolomnya

```python
filtered_delivered_approved_notna = orders_df[(orders_df['order_status']=='delivered') & orders_df['order_approved_at'].notna()].copy()
filtered_delivered_approved_notna['time_approval'] = filtered_delivered_approved_notna['order_approved_at'] - filtered_delivered_approved_notna['order_purchase_timestamp']
mean_time_approval = filtered_delivered_approved_notna['time_approval'].mean()
orders_df.loc[orders_df['order_status'] == 'delivered', 'order_approved_at'] = orders_df.loc[orders_df['order_status'] == 'delivered', 'order_approved_at'].fillna(orders_df['order_purchase_timestamp'] + mean_time_approval)
```

```python
orders_df[(orders_df['order_status']=='delivered') & orders_df['order_approved_at'].isna()].info()
```

```python
orders_df.isna().sum()
```

- mengisi beberapa kolom missing values (order_delivered_carrier_date) dengan metode Imputation pada order yang berhasil (order_status = delivered) dengan menggunakan nilai rata rata pada setiap kolomnya

```python
filtered_delivered_carrier_notna = orders_df[(orders_df['order_status']=='delivered') & orders_df['order_delivered_carrier_date'].notna()].copy()
filtered_delivered_carrier_notna['time_goods_start_shipping'] = filtered_delivered_carrier_notna['order_delivered_carrier_date'] - filtered_delivered_carrier_notna['order_approved_at']
mean_time_goods_start_shipping = filtered_delivered_carrier_notna['time_goods_start_shipping'].mean()
orders_df.loc[orders_df['order_status'] == 'delivered', 'order_delivered_carrier_date'] = orders_df.loc[orders_df['order_status'] == 'delivered', 'order_delivered_carrier_date'].fillna(orders_df['order_approved_at'] + mean_time_goods_start_shipping)
```

```python
rfm_df = all_df.groupby(by="customer_id", as_index=False).agg({
    "order_purchase_timestamp": "max", # mengambil tanggal order terakhir
    "order_id": "nunique", # menghitung jumlah order
    "payment_value": "sum" # menghitung jumlah transaksi yang dihasilkan
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```
